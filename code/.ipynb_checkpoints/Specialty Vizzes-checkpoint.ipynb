```python
sns.set(font_scale=0.9, rc={'axes.facecolor':'lightgrey', 'figure.facecolor':'darkgrey'})

pre_top5 = sns.barplot(data=precovid.sort_values('total', ascending=False).head(5),
                   x='specialty',
                   y='total',
                   palette='Purples_r',
                   saturation=0.6)

pre_top5.set(title='Top 5 Specialties (pre-covid)',
                  xlabel=None, ylabel=None)

for p in pre_top5.patches:
    height = p.get_height()
    pre_top5.text(x = p.get_x() + (p.get_width()/2),
                       y = height / 2, s = '{:.0f}'.format(height),
                       color = 'black', ha = 'center')

plt.show()
```

```python
sns.set(font_scale=0.9, rc={'axes.facecolor':'lightgrey', 'figure.facecolor':'darkgrey'}, )

pre_sat_vs_comp = sns.relplot(data=main,
                   x='satisfaction_score',
                   y='compensation_score',
                   hue='satisfaction_score',
                   s=75,
                   palette='BuGn',
                   legend=False)

pre_sat_vs_comp.set(title='Satisfaction Across Compensation (pre-covid)',
                  xlabel='Satisfaction', ylabel='Compensation')

output2 = []
for x in main.sort_values('satisfaction_score', ascending=False).head(3).itertuples():
    output2.append(x[2])

for y in output2:
    plt.text(x=main.satisfaction_score[main.specialty==y]+0.5,
         y=main.compensation_score[main.specialty==y]+0.5,
         s=y,
         fontdict=dict(color='lightblue', size=10),
         bbox=dict(facecolor='darkblue', alpha=0.85))  

output1 = []
for i in main.sort_values('compensation_score', ascending=False).head(3).itertuples():
    output1.append(i[2])

for j in output1:
    plt.text(x=main.satisfaction_score[main.specialty==j]+0.5,
         y=main.compensation_score[main.specialty==j]+0.5,
         s=j,
         fontdict=dict(color='lightgreen', size=10),
         bbox=dict(facecolor='darkgreen', alpha=0.85))  
    
plt.show()
```

```python
sns.set(font_scale=0.9, rc={'axes.facecolor':'lightgrey', 'figure.facecolor':'darkgrey'}, )

post_sat_vs_comp = sns.relplot(data=main,
                   x='post_sat_score',
                   y='post_comp_score',
                   hue='post_sat_score',
                   s=75,
                   palette='BuGn',
                   legend=False)

post_sat_vs_comp.set(title='Satisfaction Across Compensation (pre-covid)',
                  xlabel='Satisfaction', ylabel='Compensation')

output2 = []
for x in main.sort_values('post_sat_score', ascending=False).head(4).itertuples():
    output2.append(x[2])

for y in output2:
    plt.text(x=main.post_sat_score[main.specialty==y]+0.5,
         y=main.post_comp_score[main.specialty==y]+0.5,
         s=y,
         fontdict=dict(color='lightblue', size=10),
         bbox=dict(facecolor='darkblue', alpha=0.85))  

output1 = []
for i in main.sort_values('post_comp_score', ascending=False).head(3).itertuples():
    output1.append(i[2])

for j in output1:
    plt.text(x=main.post_sat_score[main.specialty==j]+0.5,
         y=main.post_comp_score[main.specialty==j]+0.5,
         s=j,
         fontdict=dict(color='lightgreen', size=10),
         bbox=dict(facecolor='darkgreen', alpha=0.85))  
    
plt.show()
```
